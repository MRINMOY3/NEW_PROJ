#### PCA 

```python
from sklearn.decomposition import PCA
```

```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
X = iris.drop('species', axis=1)
```

```python
model = PCA(n_components=2)
```

```python
model.fit(X)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
new_X = model.transform(X)

new_X = pd.DataFrame(new_X, columns=['C1', 'C2'])
```

```python
iris.columns
```

```python
sns.scatterplot('sepal_length', 'sepal_width', data=iris, hue='species')
plt.show()
```

```python
sns.scatterplot('petal_length', 'petal_width', data=iris, hue='species')
plt.show()
```

```python
plt.figure(dpi=100)
sns.scatterplot(new_X['C1'], new_X['C2'], hue=iris['species'])
plt.show()
```

```python
from sklearn.svm import SVC
```

```python
m1 = SVC()
m2 = SVC()
```

```python
m1.fit(X, iris['species'])
```

```python
m2.fit(new_X, iris['species'])
```

```python
from sklearn.metrics import accuracy_score
```

```python
p1 = m1.predict(X)
p2 = m2.predict(new_X)

y = iris['species']
```

```python
print("Accuracy: ", accuracy_score(y, p1)) # ? 
```

```python
print("Accuarcy: ", accuracy_score(y, p2))  # 2 component -> fast execute -> 
```
